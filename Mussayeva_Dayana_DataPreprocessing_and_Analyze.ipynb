```python
#importing all necessary libraries
import pandas as pd
import warnings
import matplotlib.pyplot as plot

warnings.filterwarnings('ignore')
```

```python
CSV_PATH = 'products.csv' #file name
df = pd.read_csv(CSV_PATH) #reading file
df = df['brand'] #selecting only column brand
df = df.apply(lambda x: x.strip()) #removing unneccesary spaces
df = df.to_frame() #converting back to dataframe
df = df.rename(columns={'brand': 'name'}) #renaming column brand to name
df
```

```python
#reading all files from different categories and appending it to one dataframe
data = pd.read_excel('accessories.xlsx')
data = data.append(pd.read_excel('bagsMen.xlsx'), ignore_index=True)
data = data.append(pd.read_excel('denim.xlsx'), ignore_index=True)
data = data.append(pd.read_excel('jackets.xlsx'), ignore_index=True)
data = data.append(pd.read_excel('kidsMen.xlsx'), ignore_index=True)
data = data.append(pd.read_excel('knitwear.xlsx'), ignore_index=True)
data = data.append(pd.read_excel('leatherGoodsMen.xlsx'), ignore_index=True)
data = data.append(pd.read_excel('outwearAndCoats.xlsx'), ignore_index=True)
data = data.append(pd.read_excel('pants.xlsx'), ignore_index=True)
data = data.append(pd.read_excel('shirts.xlsx'), ignore_index=True)
data = data.append(pd.read_excel('tShirts.xlsx'), ignore_index=True)
data = data.append(df, ignore_index=True)
data
```

```python
data = data.drop_duplicates() #dropping duplicate rows
data
```

```python
df = pd.read_csv('final.csv') #reading file
df
```

```python
df = df[['Country','Price']] #take out only two columns: Country and Price
df
```

```python
df['Price'] = df['Price'].str.replace('\xa0','') #removing space unicode characters from price
df['Country'] = df['Country'].astype(str).str.replace("Made in ","") #removing substring Made in from country
df.head(5) #showing top 5 rows
```

```python
df['Price'] = df['Price'].astype('int') #indicating Price column as int type
```

```python
sumValue = df.groupby(['Country']).sum() #grouping columns and showing sum of items
sumValue
```

```python
avgValue = df.groupby(['Country']).mean() #grouping columns and showing average value of items
avgValue
```

```python
avgValue.plot.bar(title="Balenciaga clothes average price per country") #plotting average price per country as bar graph 
```

```python
sumValue.plot.bar(title="Balenciaga clothes sum price per country") #plotting sum price per country as bar graph
```
